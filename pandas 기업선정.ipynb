```python
import pandas as pd
import numpy as np
import time
```

```python
def corp_select(corp_name_list):
    
    for corp_name in corp_name_list:
        pharma = pd.read_csv('./categories/'+corp_name+'.csv',encoding='utf-8')

        pharma= pharma.drop(['Unnamed: 0'],axis=1)

        pharma.sort_values('reg_day',ascending=True)

        pharma['reg_day']=pharma['reg_day'].map(lambda x : time.strptime(x, '%Y/%m/%d')[0])

        pharma_day = pharma.query('reg_day<2012')

        pharma_selected = pharma_day.sort_values(by='market_cap', ascending=False).head()

        pharma_selected=pharma_selected.reset_index().drop('index', axis=1)

        pharma_selected.to_csv('./categories/'+corp_name+'기업선정'+'.csv', encoding='utf-8')
```

```python
corp_name_list = ['제약','생명과학도구및서비스','건축자재', '생물공학', '가스유틸리티']

corp_select(corp_name_list)
```

```python
corp_concat=[]
for corp_name in corp_name_list:
    corp_concat.append(pd.read_csv('./categories/'+corp_name+'기업선정'+'.csv', encoding='utf-8'))

selected_corp = pd.concat(corp_concat).reset_index()
selected_corp.iloc[:,2:].to_csv('./categories/기업선정.csv')
```
