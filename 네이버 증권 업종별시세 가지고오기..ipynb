```python
table =pd.read_html('https://finance.naver.com/sise/sise_group_detail.nhn?type=upjong&no=35',header=0, encoding='euc-kr')
```

```python
df = table[2]
```

```python
df = df.iloc[1:,:-3]
```

```python
df.reindex
df
```

```python
df = df.dropna()
```

```python
name = df['종목명']
```

```python
new_name=[]
for i in range(1,len(name)+1):
    if i==0:
        continue
    if name[i][-1] == '*':
        new_name.append(name[i][:-2])
    else:
        new_name.append(name[i])
df['종목명']=new_name    
```

```python
df['전일대비'] = df['전일대비'].astype('f')
```
